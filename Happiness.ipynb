```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn import datasets
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=DeprecationWarning)
np.random.seed(2304)
pd.set_option('display.max_rows', 500)
```

```python
happy = pd.read_csv(r'2016.csv' )
happy
```

```python
list = ('Western Europe', 'Central and Eastern Europe')
europe=happy.loc[happy['Region'].isin(list)]
europe.shape
```

```python
def label_region (row):
   if row['Region'] == 'Western Europe' or row['Region'] == 'Central and Eastern Europe':
      return '1'
   return '0'
```

```python
happy.apply (lambda row: label_region (row),axis=1)
happy['Region_Label'] = happy.apply (lambda row: label_region (row),axis=1)
happy
```

```python
X = happy.ix[:, 6:7]
y = happy.ix[:, 3:4]
X
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X,y)
```

```python
y_proba = reg.predict(X)
```

```python
f = plt.figure(figsize=(10, 6))
plt.plot(X, y_proba[:, 0], "g-", label="Simple Regression")
plt.plot(X, y, 'bo', markersize=3)
plt.legend(loc='best')
```

```python
reg.score(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
reg2 = linear_model.LinearRegression()
reg2.fit(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

from yellowbrick.regressor import ResidualsPlot
from yellowbrick.regressor import PredictionError

ridge = Ridge()
lasso = Lasso()
visualizer = PredictionError(lasso, hist=False)

visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
g = visualizer.poof()                # Draw/show/poof the data
```
